# Performing principal component analysis on breast cancer data

I performed a Principal Component Analysis on the scikit-learn breast cancer winsconsin dataset.I load the data and target as X and y arrays, plot the original X data, train the 
PCA model, and transform the X into PCA domains. It finally displays the transformed X in PCA domains using the first two PCA components, PCA1 and PCA2. We can clearly see the differences of original data and PCA-transformed data.
Although PCA cannot be used directly for classification, it is commonly used as a dimension reduction technique before classification.

# Importing Libraries

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn import decomposition
from sklearn.datasets import load_breast_cancer
import warnings
warnings.filterwarnings("ignore")
```

```python
#Load breast cancer dataset
cancer = load_breast_cancer()
X = cancer.data
y = cancer.target
```

```python
#Plot mean radius against mean compactness
f = plt.figure(1)
plt.scatter(X[: , 0], X[: , 1], c=y)
plt.xlabel("mean radius")
plt.ylabel("mean compactness")
plt.title("Original Data")
f.show()
```

```python
#Perform PCA
pca = decomposition.PCA(n_components=3)
pca.fit(X)
X1 = pca.transform(X)
```

```python
#Plot PCA data
g = plt.figure(2)
plt.scatter(X1[:, 0], X1[:, 1], c=y)
plt.xlabel("PCA1")
plt.ylabel("PCA2")
plt.title("PCA Data")
g.show()
```
